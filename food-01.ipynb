```python
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt; plt.rcdefaults()
data = pd.read_csv('food_coded.csv')
data.head(1)
```

```python
comfort_reasons = data['comfort_food_reasons_coded'].value_counts()
print(comfort_reasons)
```

```python
import matplotlib 
matplotlib.rc('xtick', labelsize=8) 
matplotlib.rc('ytick', labelsize=8) 
objects = ('stress', 'boredom', 'depression','hunger', 'laziness', 'cold','happiness','tv','none')
y_pos = np.arange(len(objects))
performance = [22,49,18,3,4, 1, 3, 1, 5]
 
plt.bar(y_pos, performance, align='center', alpha=0.5, color='green')
plt.xticks(y_pos, objects)
plt.ylabel('Frequencies')
plt.title('Most common reason for eating')
 
plt.show()
```

```python
eating_changes = data['eating_changes_coded'].value_counts()
print(eating_changes)
```

```python
matplotlib.rc('xtick', labelsize=12) 
matplotlib.rc('ytick', labelsize=12) 
objects = ('worse', 'better', 'the same', 'unclear')
y_pos = np.arange(len(objects))
performance = [75, 36, 11, 3]
 
plt.bar(y_pos, performance, align='center', alpha=0.5, color='crimson')
plt.xticks(y_pos, objects)
plt.ylabel('Frequencies')
plt.title('Eating changes since college')
 
plt.show()
```
